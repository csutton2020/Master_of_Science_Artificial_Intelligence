```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

```python
 # change color channels order for matplotlib
# From OReily Practial computer vision
def plot_cv_img(input_image):     
    """     
    Converts an image from BGR to RGB and plots     
    """   
    # change color channels order for matplotlib     
    plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))          

    # For easier view, turn off axis around image     
    plt.axis('off')  
    plt.show()
```

```python
# Resizing and normalizing

img1 = cv2.imread("BikeShot1.jpg", cv2.IMREAD_COLOR)  # uint8 image
img2 = cv2.imread("BikeShot2.jpg", cv2.IMREAD_COLOR)  # uint8 image
img3 = cv2.imread("BikeShot3.jpg", cv2.IMREAD_COLOR)  # uint8 image
img4 = cv2.imread("BikeShot4.png", cv2.IMREAD_COLOR)  # uint8 image
output=[None, None, None, None]
source = [img1, img2, img3, img4]
for i in range(4):
    yoloSize = 416    
    dsize = (yoloSize, yoloSize)
 # resize image
    output[i] = cv2.normalize(cv2.resize(source[i], dsize, interpolation = cv2.INTER_AREA), None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
```

```python
plot_cv_img(source[0])
plot_cv_img(output[0])
plot_cv_img(source[1])
plot_cv_img(output[1])
plot_cv_img(source[2])
plot_cv_img(output[2])
plot_cv_img(source[3])
plot_cv_img(output[3])
```
