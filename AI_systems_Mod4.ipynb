```python
import nltk
# nltk.download('punkt')
# nltk.download('wordnet')
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem.wordnet import WordNetLemmatizer
wnl = WordNetLemmatizer()
import pandas as pd

```

```python
# dataframe of summary column
df = pd.read_csv('C:/Users/physi/Desktop/EN.705.603_Creating_AI_EnabledSystems/Mod4/archive/Musical_instruments_reviews.csv', usecols=["summary"])
```

```python
# word, casual, Sentence tokens printed

# word tokens separate words that use apostrophe (contractions and plural words) where casual tokens do not have the effect. 

w=[]
c=[]
s=[]
for x in df.iloc[:,0]:
    w_token = nltk.word_tokenize(x)
    w.extend(w_token)
    c_token = nltk.casual_tokenize(x)
    c.extend(c_token)
    s_token = nltk.sent_tokenize(x)
    s.extend(s_token)
print("word_tokens: " , w)
print("\n")
print("casual_tokens ", c)
print("\n")
print("sent_tokens: ", s)

```

```python
# word --> (Porter, Lancaster , Snowball) stemmers

# It seem the various stemmers have varying success of finding the root of a word. They all suffer from breaking 
# up words that have apostrophes

p=[]
l=[]
s=[]
for i in w:
    ps= PorterStemmer().stem(i)
    ls= LancasterStemmer().stem(i)
    ss=SnowballStemmer('english').stem(i)
    p.append(ps)
    l.append(ls)
    s.append(ss)
zp= dict(zip(w,p))
zl= dict(zip(w,l))
zs= dict(zip(w,s))
print("PorterSemmer: " ,zp)
print("\n")
print("Lancaster: ",zl)
print("\n")
print("SnowballStemmer: " , zs)
```

```python
# word---> lemma
#lemma seem to return the word it consumed execpt in some cases such as ex: (looking: look), (supposed: suppose) (expected: expect)

l=[]
for i in w:
    l.append(wnl.lemmatize(i,'v'))
zlem= dict(zip(w,l))
print("word_lemmatize: ",zlem)
```
