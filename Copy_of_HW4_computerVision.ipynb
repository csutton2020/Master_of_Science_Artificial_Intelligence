# Build Auto Encoder Models
Here I build model consisting of 120 Node, 60 Node, 30 Node, 10 Node and 5 Node


```python
# 120 node hidden
class AutoEncoderNeuralNetwork120(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 120),
            nn.ReLU(),
            nn.Linear(120, 120),
            nn.ReLU(),
            nn.Linear(120, 28*28),

        )

    def forward(self, x):
        x = self.flatten(x)
        solution = self.linear_relu_stack(x)
        return solution
```

```python
# 60 node hidden
class AutoEncoderNeuralNetwork60(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 60),
            nn.ReLU(),
            nn.Linear(60, 60),
            nn.ReLU(),
            nn.Linear(60, 28*28),

        )

    def forward(self, x):
        x = self.flatten(x)
        solution = self.linear_relu_stack(x)
        return solution
```

```python
# 30 node hidden
class AutoEncoderNeuralNetwork30(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 30),
            nn.ReLU(),
            nn.Linear(30, 30),
            nn.ReLU(),
            nn.Linear(30, 28*28),

        )

    def forward(self, x):
        x = self.flatten(x)
        solution = self.linear_relu_stack(x)
        return solution
```

```python
# 10 node hidden
class AutoEncoderNeuralNetwork10(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 10),
            nn.ReLU(),
            nn.Linear(10, 10),
            nn.ReLU(),
            nn.Linear(10, 28*28),

        )

    def forward(self, x):
        x = self.flatten(x)
        solution = self.linear_relu_stack(x)
        return solution
```

```python
# 5 node hidden
class AutoEncoderNeuralNetwork5(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 5),
            nn.ReLU(),
            nn.Linear(5, 5),
            nn.ReLU(),
            nn.Linear(5, 28*28),

        )

    def forward(self, x):
        x = self.flatten(x)
        solution = self.linear_relu_stack(x)
        return solution
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using {device} device")

model120 = AutoEncoderNeuralNetwork120().to(device)

# create an optimizer object
optimizer120 = optim.Adam(model120.parameters(), lr=1e-3)

# mean-squared error loss
criterion = nn.MSELoss()
```

# Train Autoencoder on MSE loss


```python
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    model.train()
    for batch, (X, y) in enumerate(dataloader):
        X, y = X.to(device), y.to(device)
        pred = model(X)
        loss = loss_fn(pred, X.view(-1,784))

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), (batch + 1) * len(X)
            print(f"loss: {loss:>7f}")
```

# 120 Node

```python
# 120 node
epochs = 20
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_dataloader, model120, criterion, optimizer120)
print("Done!")
```

```python
def test(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    test_loss, correct = 0, 0
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device)
            pred = model(X)
            test_loss += loss_fn(pred, X.view(-1,784)).item()
            # correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    # correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
epochs = 1
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    test(test_dataloader, model120, criterion)
print("Done!")
```

# Create helper to find image digit "3" in the data set

```python
def find3(tloader):
  img=None
  for train_features, train_labels in iter(tloader):
    if  train_labels[0] == 3:
       img = train_features[0].to(device)
  return img

```
